```python
from sklearn.preprocessing import LabelEncoder

encoders = {}
df_encoded = df.copy()
for i in df_encoded.select_dtypes(include='object'):
    encoder = LabelEncoder()
    df_encoded[i] = encoder.fit_transform(df_encoded[i])
    encoders[i] = encoder

df_encoded.sample(10) 
```

```python
X = df_encoded.drop(columns='time_to_rage')
y = df_encoded['time_to_rage']
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_scale = scaler.fit_transform(X)
X_scale
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X_scale,y,random_state=30,test_size=0.2)
```

```python
from lazypredict.Supervised import LazyRegressor
reg = LazyRegressor(ignore_warnings=True)
models,predictions = reg.fit(X_train,X_test,y_train,y_test)
models
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression( )
model.fit(X_train,y_train)
model.score(X_test,y_test)
```

```python
y_pred = model.predict(X_test)
y_pred
```

```python
from sklearn.metrics import root_mean_squared_error, mean_squared_error, r2_score

rmse = root_mean_squared_error(y_test,y_pred)
mse = mean_squared_error(y_test,y_pred)
r2 = r2_score(y_test,y_pred)

print(f'RMSE = {rmse}')
print(f'MSE = {mse}')
print(f'R2 Score = {r2}')
```
